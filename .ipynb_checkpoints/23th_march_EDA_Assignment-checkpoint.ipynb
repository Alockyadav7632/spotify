```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("spotify.csv")
```

### 1.Load the dataframe and ensure data quality by checking for missing values and duplicate rows.
 Handle missing values and remove duplicate rows if necessary.


```python

# Display the first few rows of the dataframe to understand its structure
print("First few rows of the dataframe:")
print(df.head())

# Check for missing values
print("\nChecking for missing values:")
print(df.isnull().sum())

# Handle missing values
# You can choose to drop rows with missing values or impute them with some values
# Here, let's drop rows with any missing values
df = df.dropna()

# Check for duplicate rows
print("\nChecking for duplicate rows:")
print(df.duplicated().sum())

# Remove duplicate rows if necessary
df = df.drop_duplicates()

# Display the shape of the dataframe after handling missing values and duplicate rows
print("\nShape of the dataframe after handling missing values and duplicates:", df.shape)
```

### 2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.


```python

# Plot histogram of popularity
plt.figure(figsize=(10, 6))
plt.hist(df['Popularity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Popularity Among Tracks')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### 3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.

```python

# Plot scatter plot of popularity vs duration
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration (ms)'], df['Popularity'], color='orange', alpha=0.5)
plt.title('Relationship between Popularity and Duration of Tracks')
plt.xlabel('Duration (milliseconds)')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

### 4.Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist
using a countplot.

```python

# Count the number of tracks for each artist
tracks_per_artist = df['Artist'].value_counts()

# Extract the artist with the highest number of tracks
artist_with_most_tracks = tracks_per_artist.idxmax()
max_tracks_count = tracks_per_artist.max()

# Display the artist with the highest number of tracks
print(f"The artist with the highest number of tracks is '{artist_with_most_tracks}' with {max_tracks_count} tracks.")

# Plot countplot of tracks per artist
plt.figure(figsize=(12, 6))
sns.countplot(y='Artist', data=df, order=df['Artist'].value_counts().index[:10], palette='viridis')
plt.title('Number of Tracks per Artist (Top 10)')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

### 5.What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.


```python

# Sort the dataframe by popularity in ascending order
df_sorted = df.sort_values(by='Popularity')

# Select the top 5 least popular tracks
top_5_least_popular = df_sorted.head(5)

# Display the artist name and track name for each of the top 5 least popular tracks
print("Top 5 Least Popular Tracks:")
for index, row in top_5_least_popular.iterrows():
    print(f"Artist: {row['Artist']} - Track: {row['Track ID']}")

```

### 6.Among the top 5 most popular artists, which artist has the highest popularity on average?
 Calculate and display the average popularity for each artist.


```python

# Group the dataframe by artist and calculate the average popularity for each artist
average_popularity_per_artist = df.groupby('Artist')['Popularity'].mean()

# Select the top 5 most popular artists
top_5_most_popular_artists = average_popularity_per_artist.nlargest(5)

# Display the average popularity for each artist
print("Average Popularity for Each Artist:")
print(top_5_most_popular_artists)

# Determine which artist among the top 5 has the highest average popularity
highest_avg_popularity_artist = top_5_most_popular_artists.idxmax()
highest_avg_popularity = top_5_most_popular_artists.max()

# Display the artist with the highest average popularity
print(f"\nAmong the top 5 most popular artists, '{highest_avg_popularity_artist}' has the highest average popularity of {highest_avg_popularity:.2f}.")
```

### 7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.

```python

# Group the dataframe by artist and calculate the maximum popularity for each artist
max_popularity_per_artist = df.groupby('Artist')['Popularity'].max()

# Sort the artists by their maximum popularity in descending order and select the top 5
top_5_most_popular_artists = max_popularity_per_artist.nlargest(5)

# For each of the top 5 most popular artists, find their most popular track
print("Most Popular Tracks for the Top 5 Most Popular Artists:")
for artist, max_popularity in top_5_most_popular_artists.items():
    artist_tracks = df[df['Artist'] == artist]
    most_popular_track = artist_tracks.loc[artist_tracks['Popularity'].idxmax(), 'Track ID']
    print(f"{artist}: {most_popular_track}")
```

### 8.Visualize relationships between multiple numerical variables simultaneously using a pair plot.

```python

# Create pair plot
sns.pairplot(df)
plt.title('Pair Plot of Numerical Variables')
plt.show()
```

### 9.Does the duration of tracks vary significantly across different artists?
Explore this visually using a box plot or violin plot.

```python

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a violin plot to visualize the distribution of track durations for each artist
sns.violinplot(x='Artist', y='Duration (ms)', data=df)
plt.title('Distribution of Track Durations Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (milliseconds)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.show()
```

### 10.How does the distribution of track popularity vary for different artists?
Visualize this using a swarm plot or a violin plot.

```python

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a violin plot to visualize the distribution of track popularity for each artist
sns.violinplot(x='Artist', y='Popularity', data=df)
plt.title('Distribution of Track Popularity Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.show()
```
